```python
import pandas as pd
```

```python
df_datos = pd.read_excel("data\Balance EEQ.xlsx", engine="openpyxl")
```

```python
print(df_datos.columns)
```

```python
df_datos = df_datos[[' Año', 'ID_Mes', 'Energía Recibida de Terceros (MWh)']]
df_datos.columns = ['Año', 'Mes', 'Energía Recibida de Terceros (MWh)']
```

```python
df_datos['Año'] = df_datos['Año'].astype("string")
df_datos['Mes'] = df_datos['Mes'].astype("string")
df_datos['Fecha'] = pd.to_datetime(df_datos['Año'] + "/" + df_datos['Mes'], yearfirst=True)
```

```python
df_datos = df_datos[['Fecha', 'Energía Recibida de Terceros (MWh)']]
df_datos['Energía Recibida de Terceros (MWh)'].fillna(0, inplace=True)
df_datos
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
def graficar_serie_temporal(df, col_x, col_y, y_label):
    _ = plt.figure(figsize=(16,6))
    _ = plt.plot(df[col_x],
                 df[col_y], 
                 marker='.', 
                 linestyle='None')

    _ = plt.ticklabel_format(style='plain', axis='y')
    _ = plt.gca().xaxis.set_major_locator(mdates.YearLocator())
    _ = plt.xlabel("Año")
    _ = plt.ylabel(y_label)
    _ = plt.xticks(rotation=45)
    _ = plt.show()
```

```python
graficar_serie_temporal(df_datos, "Fecha", "Energía Recibida de Terceros (MWh)", y_label='Energía Recibida de Terceros (MWh)')
```
